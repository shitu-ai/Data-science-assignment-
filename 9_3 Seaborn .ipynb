```python
Scatter Plot:
Use: Visualize the relationship between two continuous variables.
Example:
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Sample data
tips = sns.load_dataset("tips")

#Create a scatter plot
sns.scatterplot(x = "total_bill", y = "tip",data=tips)
plt.xlabel("Total Bill($)")
plt.ylabel("Tip($)")
plt.title("Scatter Plot:Total Bill vs. Tip")
plt.show()
```

```python
Bar Plot:
Use: Compare categorical data.
Example
```

```python
#Creata a bar plot
sns.barplot(x="day",y ="total_bill",data =tips)
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.title("Average Total Bill by Day")
plt.show()
```

```python
Histogram:
Use: Display the distribution of a single variable.
Example
```

```python
#Create a histogram
sns.histplot(tips["total_bill"],bins=20,kde=True)
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.title("Distribution of Total Bill")
plt.show()
```

```python
Heatmap:
Use: Visualize the correlation matrix of numeric variables.
Example
```

```python
#Compute correlation matrix
corr_matrix = tips.corr()

#Create a heatmap
sns.heatmap(corr_matrix,annot = True,cmap = "coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
Pair Plot:
Use: Explore pairwise relationships across multiple variables.
Example
```

```python
#Create a pair plot
sns.pairplot(tips,hue = "sex",markers=[ "o", "s"])
plt.title("Pair Plot:Tips Dataset")
plt.show()
```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Load the fMRI dataset
fmri = sns.load_dataset('fmri')

#Create a line plot
sns.lineplot(x ="timepoint",y = "signal",hue = "event",style = "region",data=fmri)
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("fMRI Signal by Timepoint ,Event , and Region")
plt.show()
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
We’ll create two box plots—one for the relationship between passenger class (pclass) and age (age), and another for the relationship between passenger class and fare (fare).
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Load the Titanic dataset
titanic_df = sns.load_dataset('titanic')

#Create box plot
plt.figure(figsize=(10,6))

#Box plot for age vs passenger class
plt.subplot(1,2,1)
sns.boxplot(x='pclass', y='age',data = titanic_df)
plt.title('Age vs Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

#Box plot for fare vs. passenger class
plt.subplot(1,2,2)
sns.boxplot(x = 'pclass',y = 'fare',data=titanic_df)
plt.title('Fare vs Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds_df = sns.load_dataset('diamonds')

# Create a histogram for price with cut as the hue
plt.figure(figsize=(8, 6))
sns.histplot(data=diamonds_df, x='price', hue='cut', kde=True, bins=30)
plt.title('Distribution of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')
plt.legend(title='Cut')

plt.show()
```

```python

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Load the flights dataset
flights = sns.load_dataset("flights")

#Create a pivot table
pt = flights.pivot("month","year","passengers")

#Create the heatmap
plt.figure(figsize=(12,8))
sns.heatmap(pt,cmap = "coolwarm",annot =True ,fmt = "d",annot_kws={"size":13})
plt.show()
```
