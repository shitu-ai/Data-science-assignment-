Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

```python
import numpy as np

# Sample data: Time spent studying (hours) and Exam scores
time_spent_studying = np.array([2, 4, 6, 8, 10])
exam_scores = np.array([50, 60, 70, 80, 90])

# Calculate the means of the two variables
mean_time = np.mean(time_spent_studying)
mean_scores = np.mean(exam_scores)

# Calculate the numerator of the Pearson correlation coefficient
numerator = np.sum((time_spent_studying - mean_time) * (exam_scores - mean_scores))

# Calculate the denominator of the Pearson correlation coefficient
denominator = np.sqrt(np.sum((time_spent_studying - mean_time) ** 2) * np.sum((exam_scores - mean_scores) ** 2))

# Calculate the Pearson correlation coefficient
r = numerator / denominator

# Print the result
print(f'Pearson correlation coefficient: {r}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data: Time spent studying (hours) and Exam scores
time_spent_studying = np.array([2, 4, 6, 8, 10])
exam_scores = np.array([50, 60, 70, 80, 90])

# Calculate the means of the two variables
mean_time = np.mean(time_spent_studying)
mean_scores = np.mean(exam_scores)

# Calculate the numerator of the Pearson correlation coefficient
numerator = np.sum((time_spent_studying - mean_time) * (exam_scores - mean_scores))

# Calculate the denominator of the Pearson correlation coefficient
denominator = np.sqrt(np.sum((time_spent_studying - mean_time) ** 2) * np.sum((exam_scores - mean_scores) ** 2))

# Calculate the Pearson correlation coefficient
r = numerator / denominator

# Print the result
print(f'Pearson correlation coefficient: {r}')

# Plot the data points
plt.scatter(time_spent_studying, exam_scores, color='blue', label='Data points')

# Plot the line of best fit
m, b = np.polyfit(time_spent_studying, exam_scores, 1)
plt.plot(time_spent_studying, m*time_spent_studying + b, color='red', label='Line of best fit')

# Adding labels and title
plt.xlabel('Time Spent Studying (hours)')
plt.ylabel('Exam Scores')
plt.title('Time Spent Studying vs Exam Scores')
plt.legend()

# Show the plot
plt.show()
```

Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.

```python
import numpy as np
import pandas as pd
from scipy.stats import spearmanr
import matplotlib.pyplot as plt

# Sample data: Sleep (hours) and Job Satisfaction
data = {
    'Sleep': [7, 6, 8, 5, 9],
    'Job_Satisfaction': [8, 7, 9, 6, 10]
}

df = pd.DataFrame(data)

# Calculate ranks
df['Rank_Sleep'] = df['Sleep'].rank()
df['Rank_Job_Satisfaction'] = df['Job_Satisfaction'].rank()

# Calculate the differences between ranks
df['d'] = df['Rank_Sleep'] - df['Rank_Job_Satisfaction']
df['d_squared'] = df['d'] ** 2

# Calculate Spearman's rank correlation coefficient
n = len(df)
sum_d_squared = df['d_squared'].sum()
spearman_rank_corr = 1 - (6 * sum_d_squared) / (n * (n**2 - 1))

# Alternatively, using scipy.stats.spearmanr for verification
spearman_corr, _ = spearmanr(df['Sleep'], df['Job_Satisfaction'])

print(f"Spearman's rank correlation coefficient (manual calculation): {spearman_rank_corr}")
print(f"Spearman's rank correlation coefficient (scipy): {spearman_corr}")

# Plot the data points
plt.scatter(df['Sleep'], df['Job_Satisfaction'], color='blue', label='Data points')

# Adding labels and title
plt.xlabel('Sleep (hours)')
plt.ylabel('Job Satisfaction (1-10)')
plt.title('Sleep vs Job Satisfaction')
plt.legend()

# Show the plot
plt.show()
```

Q3. Suppose you are conducting a study to examine the relationship between the number of hours of
exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables
for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation
between these two variables and compare the results.

```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, spearmanr

# Sample data: Exercise hours per week and BMI
data = {
    'Exercise_Hours': [3, 5, 2, 7, 4, 6, 1, 8, 9, 10],
    'BMI': [25.0, 22.5, 28.0, 20.0, 24.0, 21.5, 30.0, 19.0, 18.5, 17.5]
}
df = pd.DataFrame(data)

#Calculate Pearson correlation coefficient
pearson_corr,_ = pearsonr(df['Exercise_Hours'],df['BMI'])

# Calculate Spearman's rank correlation coefficient
spearman_corr, _ = spearmanr(df['Exercise_Hours'], df['BMI'])

print(f"Pearson correlation coefficient: {pearson_corr}")
print(f"Spearman's rank correlation coefficient: {spearman_corr}")

# Plot the data points
import matplotlib.pyplot as plt

plt.scatter(df['Exercise_Hours'], df['BMI'], color='blue', label='Data points')

# Adding labels and title
plt.xlabel('Exercise Hours per Week')
plt.ylabel('BMI')
plt.title('Exercise Hours vs BMI')
plt.legend()

# Show the plot
plt.show()
```

Q4. A researcher is interested in examining the relationship between the number of hours individuals
spend watching television per day and their level of physical activity. The researcher collected data on
both variables from a sample of 50 participants. Calculate the Pearson correlation coefficient between
these two variables.

```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr

# Sample data: TV hours per day and Physical Activity hours per day
data = {
    'TV_Hours': [2, 3, 1, 4, 2, 3, 5, 1, 4, 2],
    'Physical_Activity': [1.5, 1.0, 2.0, 0.5, 1.7, 1.2, 0.4, 2.5, 0.8, 1.8]
}

df = pd.DataFrame(data)

# Calculate Pearson correlation coefficient
pearson_corr, _ = pearsonr(df['TV_Hours'], df['Physical_Activity'])

print(f"Pearson correlation coefficient: {pearson_corr}")

# Plot the data points
import matplotlib.pyplot as plt

plt.scatter(df['TV_Hours'], df['Physical_Activity'], color='blue', label='Data points')

# Adding labels and title
plt.xlabel('TV Hours per Day')
plt.ylabel('Physical Activity Hours per Day')
plt.title('TV Hours vs Physical Activity')
plt.legend()

# Show the plot
plt.show()
```

Q5. A survey was conducted to examine the relationship between age and preference for a particular
brand of soft drink. The survey results are shown below: data = { 'Ages': [25,42 ,37,19,31,28] , 'Soft drink Preference':[Coke,Pepsi,Mountain dew,Coke,Pepsi,Coke]}


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Ages': [25, 42, 37, 19, 31, 28],
    'Soft drink Preference': ['Coke', 'Pepsi', 'Mountain dew', 'Coke', 'Pepsi', 'Coke']
}

df = pd.DataFrame(data)

# Create a bar plot to visualize the distribution of soft drink preferences among different age groups
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Soft drink Preference', hue='Ages', palette='Set2')
plt.title('Soft Drink Preference by Age')
plt.xlabel('Soft Drink Preference')
plt.ylabel('Count')
plt.legend(title='Ages', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Chi-Square Test of Independence:

To statistically test the relationship between age and soft drink preference, we can use the Chi-square test of independence. This test helps determine if there is a significant association between two categorical variables.

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['Ages'], df['Soft drink Preference'])

# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")

if p < 0.05:
    print("There is a significant association between age and soft drink preference.")
else:
    print("There is no significant association between age and soft drink preference.")
```

Q6. A company is interested in examining the relationship between the number of sales calls made per day
and the number of sales made per week. The company collected data on both variables from a sample of
30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.

```python
import numpy as np

# Hypothetical data (sales calls per day and sales per week)
sales_calls_per_day = np.array([30, 25, 40, 35, 45, 20, 30, 35, 50, 40,
                                45, 35, 25, 30, 35, 40, 30, 25, 20, 45,
                                30, 35, 40, 50, 45, 35, 25, 30, 35, 40])

sales_per_week = np.array([5, 4, 7, 6, 8, 3, 5, 6, 9, 7,
                           8, 6, 4, 5, 6, 7, 5, 4, 3, 8,
                           5, 6, 7, 9, 8, 6, 4, 5, 6, 7])

# Calculate the means of both variables
mean_sales_calls = np.mean(sales_calls_per_day)
mean_sales_week = np.mean(sales_per_week)

# Calculate the covariance between the two variables
covariance = np.cov(sales_calls_per_day, sales_per_week, ddof=0)[0, 1]

# Calculate the standard deviations of both variables
std_sales_calls = np.std(sales_calls_per_day, ddof=0)
std_sales_week = np.std(sales_per_week, ddof=0)

# Calculate Pearson correlation coefficient
pearson_corr = covariance / (std_sales_calls * std_sales_week)

print(f"Pearson correlation coefficient: {pearson_corr}")
```
