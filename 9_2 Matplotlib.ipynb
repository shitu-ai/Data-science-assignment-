```python
What is Matplotlib?
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used for data visualization in the scientific and engineering communities. Matplotlib provides a flexible and powerful way to create a wide range of plots and charts, allowing users to customize virtually every aspect of their visual representations.
```

```python
Why is it used?
Matplotlib is used for several reasons:

Versatility: It supports various types of plots and can handle multiple data formats, making it suitable for different kinds of data visualization tasks.
Customization: It offers extensive customization options, enabling users to fine-tune the appearance of their plots to meet specific requirements.
Integration: Matplotlib integrates well with other Python libraries such as NumPy, Pandas, and SciPy, making it a valuable tool in the data analysis and machine learning workflow.
Interactivity: It can create interactive plots that can be embedded in graphical user interfaces (GUIs) and web applications.
Publication Quality: It allows users to produce publication-quality figures in various formats, including PNG, PDF, and SVG.
```

```python
Line Plot:
    Line plots are used to display data points connected by straight lines. They are useful for visualizing trends over time or ordered categories.
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
Bar Plot:
    Bar plots represent data with rectangular bars. The length of each bar is proportional to the value it represents. They are useful for comparing categorical data.
    
```

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]
plt.bar(categories, values)
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
3D Plot: 
    Matplotlib also supports 3D plotting, which can be particularly useful for visualizing data with three dimensions or creating complex visual representations. 
```

```python
3D Scatter Plot
A 3D scatter plot shows data points in three-dimensional space, helping to identify patterns and relationships between three variables.
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Generate random data
x = np.random.standard_normal(100)
y = np.random.standard_normal(100)
z = np.random.standard_normal(100)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='r', marker='o')

# Set labels
ax.set_title("3D Scatter Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

plt.show()
```

```python
3D Surface Plot:
    A 3D surface plot displays a three-dimensional surface that can help visualize complex functions or data over a continuous range.
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Generate data for a 3D surface
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Add a color bar
fig.colorbar(surf)

# Set labels
ax.set_title("3D Surface Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

plt.show()
```

```python
Contour Plot::
    Contour plots are a way to represent three-dimensional data in two dimensions using contour lines. Each contour line represents a constant value, making it easy to see the gradients and slopes of the data.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Part 1: Basic Contour Plot
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.contour(X, Y, Z, levels=10, cmap='viridis')
plt.title("Contour Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar()  # Add a color bar

# Part 2: Filled Contour Plot
plt.subplot(1, 2, 2)
plt.contourf(X, Y, Z, levels=10, cmap='viridis')
plt.title("Filled Contour Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar()  # Add a color bar

plt.tight_layout()
plt.show()
```

```python
polar plot:
    Polar plots are useful for displaying data in polar coordinates, where each point is determined by an angle and a distance from the origin. Matplotlib provides functionality to create various types of polar plots.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
theta = np.linspace(0, 2 * np.pi, 100)  # Angles
r = np.abs(np.sin(theta))  # Radii

# Create a polar plot
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, projection='polar')
ax.plot(theta, r)

# Set title
plt.title("Basic Polar Plot")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
theta = 2 * np.pi * np.random.rand(100)  # Random angles
r = np.random.rand(100)  # Random radii

# Create a polar scatter plot
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, projection='polar')
ax.scatter(theta, r, c=r, cmap='viridis', alpha=0.75)

# Set title
plt.title("Polar Scatter Plot")

plt.show()
```

```python
A scatter plot is a type of data visualization that displays values for typically two variables for a set of data. The values of one variable determine the position on the horizontal axis (x-axis), and the values of the other variable determine the position on the vertical axis (y-axis). Scatter plots are used to observe relationships between variables and to identify patterns or trends.
```

```python
import matplotlib.pyplot as plt

#Generate data for x and y
np.random.seed(3)
x = 3+np.random.normal(0,2,50)
y = 3+np.random.normal(0,2,len(x))

#Create the scatter plot
plt.scatter(x,y,color ='b', marker = 'o',label = 'Data Points')
plt.title("Scatter Plot:Relationship between x and y ")
plt.xlabel("x")
plt.ylabel("y")

#Add a trend line(line of best fit
coefficients = np.polyfit(x,y,1)
trend_line = np.poly1d(coefficients)
plt.plot(x,trend_line(x),color = 'r', label = 'Trend Line')

plt.legend()
plt.grid(True)
plt.show()
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
The subplot() function in Matplotlib allows you to create multiple plots within the same figure. Its particularly useful when you want to compare different datasets or visualize related information side by side. Lets create four line plots using the provided data:
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line plots
x1, y1 = np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500])
x2, y2 = np.array([0, 1, 2, 3, 4, 5]), np.array([50, 20, 40, 20, 60, 70])
x3, y3 = np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60])
x4, y4 = np.array([0, 1, 2, 3, 4, 5]), np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title("Line 4")

plt.tight_layout()  # Adjust subplot spacing
plt.show()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
A bar plot, also known as a bar chart, visually represents data using rectangular bars of varying heights. Its commonly used to compare numerical data across different categories or groups. The bars can be oriented either horizontally or vertically. Here are the key points about bar plots:

Definition:
A bar plot displays data using rectangular bars, where each bar corresponds to a category, and the length of the bar represents the value of the data.
Bar graphs allow easy comparison of quantities and reveal patterns in the data1.
Uses:
Comparing data across different categories or groups.
Showing changes over time.
Representing frequency distributions.
Highlighting relationships between variables.
Creating a Bar Plot:
Lets create both a vertical bar plot and a horizontal bar plot for the given data:
```

```python
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company Profits")
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='salmon')
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Company Profits")
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
A box plot, also known as a box-and-whisker plot, is a graphical representation of data that summarizes its distribution using five key summary statistics:

Minimum: The smallest value in the dataset.
First Quartile (Q1): The median of the lower half of the data.
Median (Q2): The middle value that divides the dataset into two equal parts.
Third Quartile (Q3): The median of the upper half of the data.
Maximum: The largest value in the dataset.
Box plots provide insights into the shape, symmetry, variability, and presence of outliers in the data distribution. They are particularly useful when comparing multiple datasets or identifying skewness.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.boxplot([box1, box2], labels=["Box 1", "Box 2"])
plt.title("Box Plot for Box1 and Box2")
plt.xlabel("Box")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```
